```python
import csv
f = open('subwayfee.csv',encoding='UTF-8')
data = csv.reader(f)
next(data)
mx = 0
rate = 0
station = ''
for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    if row[6] != 0 and (row[4]+row[6])> 100000:
        rate = row[4]/(row[4]+row[6])
        if rate>mx:
            mx = rate
            station = row[3]
            
print(station,mx)
```

```python
import csv
f = open('subwayfee.csv',encoding='UTF-8')
data = csv.reader(f)
next(data)

mx = [0]*4
station = ['']*4

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
        if row[i] > mx[i-4]:
            mx[i-4] = row[i]
            station[i-4] = row[3]+' '+row[1]
print(station)
print(mx)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv',encoding='UTF-8')
data = csv.reader(f)
next(data)
label = ['유임승차','유임하차','무임승차','무임하차']

plt.rc('font',family = 'Malgun Gothic')

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
    if "목동" in row[3]:

        plt.title(row[3]+' ' +row[1])
        plt.pie(row[4:8],labels=label,autopct='%1.f%%')
        plt.axis('equal')
        
        #현재 이미지를 저장해주는 명령어
        plt.savefig(row[3]+' '+row[1]+'.png')
        plt.show()
```
