```python
import numpy as np
import matplotlib.pyplot as plt
```

### 삼각 함수 사용

```python
print(np.pi)
print(np.sin(0))
print(np.cos(np.pi))
```

### 랜덤 함수 사용

```python
a = np.random.rand(5)
b = np.random.randint(5)
print(a)    #배열 형식으로 생성
print(b)    #정수 형식으로 생성

#0부터 5까지 정수 10개 뽑기
print(np.random.choice(6,10,replace=True))    #replace를 False로 하면 중복 값 선택 안함
#속성 p를 사용하여 각각 뽑힐 확률 지정
print(np.random.choice(6,10,p=[0.1,0.2,0.3,0.2,0.1,0.1]))
```

## numpy를 활용하여 그래프 그리기

### 히스토그램

```python
dice = np.random.choice(6,10000)
print(dice)

plt.hist(dice,bins=6)
plt.show()
```

### 버블 차트

```python
x = np.random.randint(10,100,200)
y = np.random.randint(10,100,200)
size = np.random.randint(10,100,200)

plt.scatter(x,y,s=size,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

## array 생성하기

### 기본 문법

```python
n = 10
#0으로 이루어진 배열 n개 생성하기
a = np.zeros(n)
print(a)

#1로 이루어진 배열 n개 생성하기
b = np.ones(n)
print(b)

# n by n 단위행렬 생성
c = np.eye(n)
print(c)
```

### arange 사용

```python
# 1부터 2까지 0.1 간격으로 생성
d = np.arange(1,2,0.1)
print(d)

# 1부터 2를 11칸 나눈 배열 생성
e = np.linspace(1,2,11)
print(e)
```

### array 활용
>ndarray를 연산이나 함수를 적용하면 모든 원소에 적용됨

```python
x = np.arange(-np.pi,np.pi,np.pi/100)
plt.plot(x,np.sin(x))
plt.plot(x+np.pi,np.sin(x))
plt.show()
```

## mask 사용

```python
a = np.arange(-5,5)

print(a)
print(a<0)
print(a[a<0])

#변수 사용하여 마스크 적용

mask1 = abs(a)>3
mask2 = abs(a)%2 ==0
print(a[mask1])

#and 조건
print(a[mask1*mask2])

#or 조건
print(a[mask1+mask2])
```

### maks 활용하기

```python
x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

#mask 설정
m1 = abs(x)>50
m2 = abs(y)>50

#mask 적용
x = x[m1+m2]
y = y[m1+m2]
size = np.random.randint(10,100,len(x))

plt.scatter(x,y,s=size,c=y,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```
