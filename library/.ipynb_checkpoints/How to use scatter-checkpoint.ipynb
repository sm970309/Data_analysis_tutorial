```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#(x축,y축,s=크기,c=색깔(range를 사용하면 각기 다른 색으로 표현 가능))
plt.scatter([1,2,3,4],[10,20,30,40],s=[100,200,150,200],c=range(4),cmap = 'jet')

plt.colorbar()
plt.show()
```

```python
import random

x=[]
y=[]
size=[]

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))

#alpha는 투명도
plt.scatter(x,y,s=size,cmap='jet',c=size,alpha=0.7)
plt.colorbar()
plt.show()
```
