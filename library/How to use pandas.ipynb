# pandas example
>read_html 사용하여 html파일 내의 자료 불러오기

```python
import pandas as pd

#html에서 코드 읽기,header와 index 정하기(1번 열에 나라 이름)
df = pd.read_html('https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/medal-standings.htm',header=0,index_col=1)

tokyo = df[0].iloc[:,:5]
```

### iloc[a:b,c:d]
> a부터b행까지, c열부터d열까지

```python
tokyo.columns = ['순위','금','은','동','계']    #열 이름 정하기
tokyo = tokyo.sort_values('금',ascending=False)    #sort_values('열 이름')로 기준 정해서 정렬

tokyo
```

## 위의 자료를 엑셀 형태로 저장
>to_excel

```python
tokyo.to_excel('도쿄올림픽메달.xlsx')
```

# 데이터 프레임 기초
### DataFrame = 2차원 형태의 데이터

```python
# date_range('일/월/년',periods=기간)
index = pd.date_range('1/1/2021',periods = 60)
print(index)
```

```python
import numpy as np
df = pd.DataFrame(np.random.rand(60,3),index = index,columns=list('ABC'))
df
```

```python
# mask 적용 가능

mask = df['B'] > 0.7

df[mask]
```

### 새로운 열 생성하기

```python
df['D'] = df['A']/df['B']

df
```

### axis = 0이면 열 기준, 1이면 행 기준

```python
df['E'] = np.sum(df,axis=1)

df.head()   #처음 5개의 데이터만 출력하는 코드
```

```python
df = df.sub(df['A'],axis = 0)
df.head()
```
