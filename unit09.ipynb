```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
size = []
name = input("찾고싶은 지역을 입력하세요: ")

for row in data:
    if name in row[0]:
        m,f=0,0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

plt.title(name+" 남녀 비율")
plt.rc('font',family='Malgun Gothic')
plt.pie(size,labels=["남자","여자"],startangle=90,autopct='%.3f%%')
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
name = input("찾고싶은 지역을 입력하세요: ")

m=[]
f=[]

for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+3]))
            f.append(int(row[i+106]))
        break

plt.plot(m,label='남',color='b')
plt.plot(f,label='여',color='r')
plt.legend()
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
name = input("찾고싶은 지역을 입력하세요: ")

result = []
for row in data:
    if name in row[0]:
        for i in range(101):
            result.append(int(row[i+3])-int(row[i+106]))
        break
plt.rcParams['axes.unicode_minus'] = False
plt.bar(range(101),result)
plt.show()
```

```python
import math
f = open('gender.csv')
data = csv.reader(f)
name = input("찾고싶은 지역을 입력하세요: ")

m=[]
f=[]
size = []
for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+3]))
            f.append(int(row[i+106]))
            size.append(math.sqrt(int(row[i+3])+int(row[i+106])))
        break
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5))
plt.title(name+"의 성별 인구 그래프")

plt.scatter(m,f,c=range(101),s=size,alpha=0.5,cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)))

plt.show()
```
