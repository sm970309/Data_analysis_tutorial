# 히스토그램으로 나타내기

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul(1904-).csv')
data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =='08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.hist(aug,bins=100,color='r')
plt.hist(jan,bins=100,color='b')

plt.show()
f.close()
```

```python
f = open('seoul(1904-).csv')
data = csv.reader(f)
next(data)

day = [[]for _ in range(31)]

for row in data:
    date = int(row[0].split('-')[2])
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            day[date-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(day)
plt.show()
f.close()
```
