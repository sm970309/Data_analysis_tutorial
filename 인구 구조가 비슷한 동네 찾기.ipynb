```python
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
next(data)

data = list(data)

name = input("알고싶은 동네를 입력해주세요: ")
minv = 1    #확률 최소값
result_name = ''    #가장 차이가 안나는 지역 이름
result = []         #가장 차이가 안나는 지역 데이터 저장

for row in data:
    if name in row[0]:
        home = np.array(row[3:],dtype = int)/int(row[2])
        break

for row in data:
    away = np.array(row[3:],dtype = int)/int(row[2])
    s = np.sum(abs(home-away))
    if s<minv and name not in row[0]:
        minv = s
        result_name = row[0]
        result = away

plt.style.use('ggplot')
plt.figure(figsize = (10,5))
plt.rc('font',family = 'Malgun Gothic')

plt.plot(home,label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```

# pandas의 DataFrame을 사용하여 분석하기

```python
import pandas as pd
```

### csv파일 불러오기

```python
df = pd.read_csv('age.csv',encoding = "cp949",index_col=0)
df.head()
```

### 모든 값을 '2021년06월 계 총인구수' 의 열 기준으로 나누기

```python
df = df.div(df['2021년06월_계_총인구수'],axis = 0)
df.head()
```

```python
#참고 (열기준,행기준)
print(np.sum(df,axis=1))    #행기준으로 더함
print(np.sum(df,axis=0))    #열기준으로 더함
```

### 필요 없는 열 삭제하기 (del)

```python
del df['2021년06월_계_총인구수'],df['2021년06월_계_연령구간인구수']
df.head()
```

### 궁금한 지역 입력받기

```python
name = input('원하는 지역 입력: ')
a = df.index.str.contains(name) #name in df 와 같은 개념
df2 = df[a]    #mask 적용
df2
```

### 전치행렬을 만든 후 plot적용

```python
df2.T.plot()
plt.show()
```

### 궁금한 지역의 인구구조와 비슷한 지역 시각화

```python
x = df.sub(df2.iloc[0],axis=1)    #iloc[0] 은 iloc[0,:]과 같다.
y = np.power(x,2)
z = y.sum(axis=1)

x.loc['경기도 송탄출장소 (4122200000)']
```

```python
i = z.sort_values()    #index를 사용하면 index만 추출할 수 있음
count = 0
while i.iloc[count]==0:
    count+=1
print('0인 지역의 수:',count)
i = i.index[count+1:count+6]
i
```

```python
df.loc[i].T.plot()
plt.show()
```
