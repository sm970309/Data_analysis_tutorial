```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)

header = next(data)
result = []

for row in data:
    if "서울특별시 양천구 신정4동" in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.bar(range(101),result)
plt.show()

plt.barh(range(101),result)
plt.show()
f.close()
```
